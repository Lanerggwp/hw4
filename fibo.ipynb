###Write code to measure the execution time of F(10), F(20), ..., F(100) using both pure recursive and dynamic programming methods. Plot the results as a line chart.


```python
import time
import matplotlib.pyplot as plt

# recursive
def Fibo_re(n):
    if n <= 1:
        return n
    return Fibo_re(n-1) + Fibo_re(n-2)

# dynamic proggraming
def Fibo_dy(n):
    f = [0,1]
    for i in range(2,n+1):
        f.append(f[i-1]+f[i-2])
    return f[n]

#recursive algo. time
recursive_time = []
dynamic_time = []
for n in range(10,101,10):
  if(n<50):
    start_time = time.time()
    Fibo_re(n)
    end_time = time.time()
    recur_time = end_time-start_time
    recursive_time.append(recur_time)
  else:
    recursive_time.append(99)

  start_time = time.time()
  Fibo_dy(n)
  end_time = time.time()
  dyna_time = end_time-start_time
  dynamic_time.append(dyna_time)

#line chart
plt.plot(range(10,101,10),recursive_time,label='recursive time')
plt.plot(range(10,101,10),dynamic_time,label='dynamic time')
plt.xlabel('n')
plt.ylabel('time')
plt.title('Execution time of recursive & dynamic method algo. of fibonacci.')
plt.legend()
plt.show()
```
